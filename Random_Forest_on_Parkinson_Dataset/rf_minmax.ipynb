```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
p=pd.read_csv('parkinson_min_max.csv')
```

```
p.head()
```

```
x_df=p.drop(['Unnamed: 0','class'],axis=1)
y_df=p['class']
x_df.head()
```

```
from sklearn.model_selection import train_test_split
X_Train, X_Test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2, random_state=17)
```

```
# Fitting the classifier into the Training set
from sklearn.ensemble import RandomForestClassifier
```

```
Classifier=RandomForestClassifier()
```

```
## Hyper Parameter tuning
n_estimators = [60,80, 100, 500, 900, 1100]
criterion=['gini','entropy']
max_depth = [10,15,20,25,30]

# Define the grid of hyperparameters to search
hyperparameter_grid = {
    'n_estimators': n_estimators,
    'max_depth':max_depth,
    'criterion':criterion
    }
```

```
from sklearn.model_selection import RandomizedSearchCV 
random_cv = RandomizedSearchCV(estimator=Classifier,
            param_distributions=hyperparameter_grid,
            cv=5, n_iter=50,
            scoring = 'neg_mean_absolute_error',n_jobs = 4,
            verbose = 5, 
            return_train_score = True,
            random_state=17)
```

```
random_cv.fit(X_Train,y_train)
```

```
random_cv.best_estimator_
```

```
Classifier = RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=20, max_features='auto',
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=82,
                       n_jobs=None, oob_score=False, random_state=None,
                       verbose=0, warm_start=False)
```

```
Classifier.fit(X_Train,y_train)
```

```
y_pred=Classifier.predict(X_Test)
```

```
y_pred
```

```
tes=np.array(y_test)
tes
```

```
np.sum(np.abs(tes-y_pred))
```

```
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,tes)
cm
```
